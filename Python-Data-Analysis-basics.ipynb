```python
kanto = [73,67,43]
weights = [0.3,0.2,0.5]
result = 0.0
for r,w in zip(kanto,weights):
    result += r*w
    
print(result)
```

```python
import numpy as np
```

```python
kanto = np.array(kanto)
weights = np.array(weights)


```

```python
regions = np.array([
    [12,13,14],
    [18,17,16],
    [21,23,27],
    [31,22,15],
    [25,19,31]
])
regions
```

```python
regions.shape
```

```python
yeilds = regions @ weights
yeilds = yeilds.reshape(5,1)
yeilds
```

```python
result = np.concatenate((regions,yeilds),axis=1)
result
```

```python
np.savetxt('yield.txt',result,fmt='%.2f',comments = '',header = 'temperature,rainfall,humidity,yield_apples')
```
